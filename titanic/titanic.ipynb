```python
#Data cleaning

df=df.drop(['PassengerId','Cabin','Ticket'],axis=1)
```

```python
df
```

```python
#cleaning null values

df.isna().sum()
```

```python
df=df.dropna(subset=['Age'])
```

```python
df
```

```python
#reformatting categorical data

df['Embarked']=df['Embarked'].replace("S","1")
df['Embarked']=df['Embarked'].replace("C","2")
df['Embarked']=df['Embarked'].replace("Q","3")
```

```python
df
```

# Exploratory Analysis

```python
#Total numbers of male that survived
total_males=((df['Sex']=='male') & (df['Survived']==1)).sum()
print(f'Total Number of male survivers: {total_males}')

# Total numbers of males that died
total_males_death=((df['Sex']=='male') & (df['Survived']==0)).sum()
print(f'Total Number of males that died: {total_males_death}')

array_males=[total_males,total_males_death]
```

```python
#Total numbers of female that survived
total_females=((df['Sex']=='female') & (df['Survived']==1)).sum()
print(f'Total Number of female survivers: {total_females}')

# Total numbers of males that died
total_females_death=((df['Sex']=='female') & (df['Survived']==0)).sum()
print(f'Total Number of females that died: {total_females_death}')

array_females=[total_females,total_females_death]
```

```python
x=['MALES']

plt.bar(x,total_males_death,color='y',label='Death',width=0.5)
plt.bar(x,total_males,color='b',bottom=total_males_death,label='total',width=0.5)
x1=['FEMALES']
plt.bar(x1,total_females_death,color='y',width=0.5)
plt.bar(x1,total_females,color='b',bottom=total_females_death,width=0.5)
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Survivors v/s Deaths")
plt.legend()
plt.show()
```

```python
sns.histplot(df[df["Survived"] > 0]["Age"], kde=True, bins=9)
plt.xlabel("Ages")
plt.ylabel("Number of Passengers")
plt.title("Age Categories v/s Number of Passengers Survived")
plt.show()
```

```python
minor_children_death=((df['Age']<16) & (df['Survived']==0)).sum()
print(f'Number of minor children death are: {minor_children_death}')

minor_children_survivors=((df['Age']<16) & (df['Survived']==1)).sum()
print(f'Number of minor children survivors are: {minor_children_survivors}')
```

```python
sns.histplot(df[df["Survived"] == 0]["Age"], kde=True, bins=9)
plt.xlabel("Ages")
plt.ylabel("Number of Passengers")
plt.title("Age Categories v/s Number of Passengers Died")
plt.show()
```

```python
#dead passangers wrt ticket price

death_df=df[df['Survived']==0]
```

```python
death_df
```

```python
sns.displot(death_df,x='Fare',binwidth=40,kde=True)
plt.xticks([0,50,100,150,200,250])
plt.title("Fare v/s People who died")
plt.show()
```

```python
total_rich_people=(df['Fare']>100).sum()
print(f"Total people who paid more than $100: {total_rich_people}")
people_survived=((df['Fare']>100) & (df['Survived']==1)).sum()
print(f"Total people who survived: {people_survived}")
print(f"Percentage of rich people survived: {(people_survived/total_rich_people)*100}%")
# clearly there has was biasness
```
